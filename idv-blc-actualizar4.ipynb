```haskell
-- AUXILIARY FUNCTIONS IMPLEMENTED

-- To join a list of strings with a comma and space
joinWithComma :: [String] -> String
joinWithComma [] = ""
joinWithComma [s] = s
joinWithComma (s:ss) = s ++ ", " ++ joinWithComma ss

-- Indentation
indent :: String -> String
indent s = unlines (map ("  " ++) (lines s))
```

#### 3. An `addToCatalog Publication Catalog` function that adds a publication to the library catalog. If the publication is already in the catalog, its number of copies will be increased. 

##### DESCRIPTION OF FUNCTIONS IMPLEMENTED

+ **Cosa implementada**: is a `Tipo`, representing the (...). 

```haskell
addToCatalog :: Publication -> Catalog -> Catalog
addToCatalog pub (Catalog entries) = Catalog (updatedEntries ++ newEntry)
  where
    existingEntry = filter (\entry -> publication entry == pub) entries
    updatedEntries = map updateEntry entries
    updateEntry entry
      | publication entry == pub = entry {copies = copies entry ++ [newCopy]}
      | otherwise = entry
      where
        nextCopyId = if null (copies entry) then 1 else 1 + maximum (map copyId (copies entry))
        newCopy = Copy nextCopyId Nothing
    newEntry
      | null existingEntry = [CatalogEntry pub [Copy 1 Nothing]]
      | otherwise = []
```

#### 4. A `removeFromCatalog Publication Catalog` that removes a copy of the publication from the library catalog. If it is the last copy, the publication will be removed. If the catalog does not contain that publication an error will be raised. 

##### DESCRIPTION TYPES CREATED / DESCRIPTION OF FUNCTIONS IMPLEMENTED

+ **Cosa implementada**: is a `Tipo`, representing the (...). 

```haskell
-- Remove a copy of the publication from the catalog
removeFromCatalog :: Publication -> Catalog -> Catalog
removeFromCatalog pub (Catalog entries) = Catalog (processEntries entries)
  where
    -- Recursive function to process entries
    processEntries :: [CatalogEntry] -> [CatalogEntry]
    processEntries [] = error "Publication not found in the catalog."
    processEntries (entry:rest)
      | publication entry == pub = 
          case removeCopy entry of
            Nothing -> rest -- Remove the entire entry if no copies remain
            Just updatedEntry -> updatedEntry : rest -- Keep the updated entry
      | otherwise = entry : processEntries rest

    -- Remove a copy from a CatalogEntry; if no copies remain, return Nothing
    removeCopy :: CatalogEntry -> Maybe CatalogEntry
    removeCopy (CatalogEntry p copies) =
      case copies of
        [] -> Nothing -- No copies remain
        [_] -> Nothing -- Remove the entire entry if it's the last copy
        (_:cs) -> Just (CatalogEntry p cs) -- Remove the first copy
```

#### 5. A `borrowedPublications User Catalog` function that returns the publications the user has borrowed. They must be sort by closest return date, in case of tie in return date, it will first show books, then journals and finally DVDs.

##### DESCRIPTION TYPES CREATED / DESCRIPTION OF FUNCTIONS IMPLEMENTED

+ **Cosa implementada**: is a `Tipo`, representing the (...). 

```haskell
-- Our implementatios for mapMaybe and sortBy
ourMapMaybe :: (a -> Maybe b) -> [a] -> [b]
ourMapMaybe _ [] = []
ourMapMaybe f (x:xs) =
  case f x of
    Just y  -> y : ourMapMaybe f xs
    Nothing -> ourMapMaybe f xs
    
ourSortBy :: (a -> a -> Ordering) -> [a] -> [a]
ourSortBy _ [] = []
ourSortBy cmp (x:xs) = insertBy cmp x (ourSortBy cmp xs)

insertBy :: (a -> a -> Ordering) -> a -> [a] -> [a]
insertBy _ x [] = [x]
insertBy cmp x (y:ys)
  | cmp x y == GT = y : insertBy cmp x ys
  | otherwise     = x : y : ys
-- Task 5
-- Function to extract borrowed publications by a specific user
borrowedPublications :: User -> Catalog -> [Publication]
borrowedPublications user (Catalog entries) =
  let
    -- Collect all borrowed copies by user
    borrowedCopies = concatMap (getCopies user) entries

    -- Extract publications and return dates
    publicationsWithDates = map (\(copy, pub) -> (pub, getReturnDate copy)) borrowedCopies

    -- Definite sort
    sortedPublications = map fst $ ourSortBy comparePublications publicationsWithDates in sortedPublications

-- Helper: Get borrowed copies by a specific user from a CatalogEntry
getCopies :: User -> CatalogEntry -> [(Copy, Publication)]
getCopies user entry = ourMapMaybe (\copy -> if isBorrowedBy user copy
                        then Just (copy, publication entry)
                        else Nothing) (copies entry)

-- Helper: Check if a Copy is borrowed by the given User
isBorrowedBy :: User -> Copy -> Bool
isBorrowedBy user (Copy _ (Just (Borrowing borrower _ _))) = user == borrower
isBorrowedBy _ _ = False

-- Helper: Extract the return date from a borrowed Copy
getReturnDate :: Copy -> Date
getReturnDate (Copy _ (Just (Borrowing _ _ returnDate))) = returnDate
getReturnDate _ = error "Copy is not borrowed"

-- Comparison function for sorting
comparePublications :: (Publication, Date) -> (Publication, Date) -> Ordering
comparePublications (pub1, date1) (pub2, date2)
  | date1 /= date2 = compare date1 date2  -- Compare by return date
  | otherwise      = comparePublicationType pub1 pub2  -- Tie-break by publication type

-- Helper: Compare publication types: Books < Papers < DVDs
comparePublicationType :: Publication -> Publication -> Ordering
comparePublicationType Book{} Book{} = EQ
comparePublicationType Book{} _      = LT  -- Books come first
comparePublicationType _ Book{}      = GT
comparePublicationType Paper{} Paper{} = EQ
comparePublicationType Paper{} _     = LT  -- Papers come second
comparePublicationType _ Paper{}     = GT
comparePublicationType DVD{} DVD{}   = EQ
```

```haskell
-- Function to get all publications by an author
publicationsByAuthor :: Person -> Catalog -> [Publication]
publicationsByAuthor author (Catalog entries) = concatMap (publifilter author) entries

-- Helper: Filter publications by author within a single CatalogEntry
publifilter :: Person -> CatalogEntry -> [Publication]
publifilter author (CatalogEntry pub _)
  | authorCheck author pub = [pub]
  | otherwise = []

-- Helper: Check if a person is an author/director/actor in a publication
authorCheck :: Person -> Publication -> Bool
authorCheck author (Book _ _ bookAuthor _) = author == bookAuthor
authorCheck author (Paper _ _ paperAuthors _ _) = author `elem` paperAuthors
authorCheck author (DVD _ _ dvdDirector dvdActors _ _) =
  author == dvdDirector || author `elem` dvdActors

-- Function to get books by an author
booksByAuthor :: Person -> Catalog -> [Publication]
booksByAuthor author (Catalog entries) = concatMap (filterBooksByAuthor author) entries

-- Helper: Filter books by author within a single CatalogEntry
filterBooksByAuthor :: Person -> CatalogEntry -> [Publication]
filterBooksByAuthor author (CatalogEntry pub _)
  | isBookByAuthor author pub = [pub]
  | otherwise = []

-- Helper: Check if a publication is a book authored by the person
isBookByAuthor :: Person -> Publication -> Bool
isBookByAuthor author (Book _ _ bookAuthor _) = author == bookAuthor
isBookByAuthor _ _ = False

-- Function to get journals (papers) by an author
journalsByAuthor :: Person -> Catalog -> [Publication]
journalsByAuthor author (Catalog entries) = concatMap (filterJournalsByAuthor author) entries

-- Helper: Filter journals by author within a single CatalogEntry
filterJournalsByAuthor :: Person -> CatalogEntry -> [Publication]
filterJournalsByAuthor author (CatalogEntry pub _)
  | isJournalByAuthor author pub = [pub]
  | otherwise = []

-- Helper: Check if a publication is a journal (paper) authored by the person
isJournalByAuthor :: Person -> Publication -> Bool
isJournalByAuthor author (Paper _ _ paperAuthors _ _) = author `elem` paperAuthors
isJournalByAuthor _ _ = False

-- Function to get DVDs by an author (considering both director and actors)
dvdsByAuthor :: Person -> Catalog -> [Publication]
dvdsByAuthor author (Catalog entries) = concatMap (filterDVDsByAuthor author) entries

-- Helper: Filter DVDs by author within a single CatalogEntry
filterDVDsByAuthor :: Person -> CatalogEntry -> [Publication]
filterDVDsByAuthor author (CatalogEntry pub _)
  | isDVDByAuthor author pub = [pub]
  | otherwise = []

-- Helper: Check if a publication is a DVD authored by the person
isDVDByAuthor :: Person -> Publication -> Bool
isDVDByAuthor author (DVD _ _ dvdDirector dvdActors _ _) =
  author == dvdDirector || author `elem` dvdActors
isDVDByAuthor _ _ = False

```
